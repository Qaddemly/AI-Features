```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.cluster import SpectralClustering
from sklearn.metrics import pairwise_kernels
import matplotlib.pyplot as plt
```

```python
Jobs = pd.read_csv("JobsFE.csv")
```

```python
Jobs.columns
```

```python
# Combine relevant columns into a single text column
Jobs["job_text"] = (
    Jobs["position"].astype(str) + " " +
    Jobs["job_role_and_duties"].astype(str) + " " +
    Jobs["requisite_skill"].astype(str)
)
```

```python
# Initialize SBERT model
MODEL = SentenceTransformer("paraphrase-MiniLM-L6-v2", device="cpu")
job_texts = Jobs["job_text"].tolist()
```

```python
# Generate SBERT embeddings
job_embeddings = MODEL.encode(job_texts, convert_to_numpy=True)
job_embeddings = job_embeddings / np.linalg.norm(job_embeddings, axis=1, keepdims=True)
```

### Elbow Method to determine the optimal number of clusters


```python
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score

def find_best_spectral_clusters(data, max_clusters=20):
    best_k = 2  # أقل عدد للعناقيد الممكنة
    best_score = -1
    scores = []

    similarity_matrix = pairwise_kernels(data, metric="cosine")

    for k in range(2, max_clusters + 1):  # نبدأ من 2 لأن k=1 لا معنى له هنا
        spectral = SpectralClustering(n_clusters=k, affinity="precomputed", random_state=42)
        cluster_labels = spectral.fit_predict(similarity_matrix)

        score = silhouette_score(data, cluster_labels)
        scores.append(score)

        if score > best_score:
            best_k = k
            best_score = score

    # رسم القيم
    plt.figure(figsize=(8, 5))
    plt.plot(range(2, max_clusters + 1), scores, "bo-", markersize=8)
    plt.xlabel("عدد العناقيد (k)")
    plt.ylabel("Silhouette Score")
    plt.title("تحليل Silhouette Score لاختيار العدد الأمثل للعناقيد")
    plt.show()

    return best_k


```

```python
# find optimal_clusters
optimal_clusters = find_best_spectral_clusters(job_embeddings, max_clusters=20)
print(f"العدد الأمثل للعناقيد: {optimal_clusters}")
```

#### Perform Spectral Clustering with the optimal number of clusters


```python
similarity_matrix = pairwise_kernels(job_embeddings, metric="cosine")
spectral = SpectralClustering(n_clusters=optimal_clusters, affinity="precomputed", random_state=42)
cluster_labels = spectral.fit_predict(similarity_matrix)
```

```python
# Add the cluster labels as a new column to the dataset
Jobs["cluster"] = cluster_labels
```

```python
# Save the updated dataset with cluster labels
Jobs.to_csv("Jobs_with_clusters_spectral.csv", index=False)
```

```python
# Encode the resume text
resume_embedding = MODEL.encode([resume_text], convert_to_numpy=True)
```

```python
# calculate the similarity between job embedding and resume embedding 
resume_similarities = pairwise_kernels(resume_embedding, job_embeddings, metric="cosine")

# push resume to nearest cluster 
resume_cluster = cluster_labels[np.argmax(resume_similarities)]

print(f"The resume belongs to cluster: {resume_cluster}")
```

#### Function to evaluate model results using Spectral cluster 

```python
def evaluate_model_results(recommended_jobs, resume_cluster):
    """
    Evaluates the model results by checking how many recommended jobs
    belong to the same cluster as the resume.
    """
    # Extract cluster labels of recommended jobs
    recommended_clusters = []
    for job in recommended_jobs:
        job_id = job["Job Id"]  # Ensure "Job Id" is the correct key
        cluster = Jobs.loc[Jobs["Job Id"] == job_id, "cluster"].values[0]
        recommended_clusters.append(cluster)

    # Calculate cluster purity
    cluster_purity = np.sum(np.array(recommended_clusters) == resume_cluster) / len(recommended_clusters)

    print(f"Clusters of recommended Jobs: {recommended_clusters}")
    print(f"Cluster of resume is {resume_cluster}")
    print(f"Percentage of recommended jobs in the same cluster: {cluster_purity:.2%}")
    return cluster_purity
```

#### make language and proficiency_levels
Implementation Plan

Convert both columns from string format to lists.

Pair each language with its corresponding proficiency level (if available).

Create a final merged languages column in the format:

```python
# Function to safely convert stringified lists to actual lists
def safe_list_conversion(val):
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)  # Convert string to list
        except:
            return []  # Handle invalid formats
    elif isinstance(val, list):
        return val  # Already a list
    return []  # Return empty list if NaN or other types

# Convert columns
resume_data["languages"] = resume_data["languages"].apply(safe_list_conversion)
resume_data["proficiency_levels"] = resume_data["proficiency_levels"].apply(safe_list_conversion)
print(resume_data["languages"].value_counts())
print(resume_data["proficiency_levels"].value_counts())
```

```python
# Standardize proficiency levels
def clean_proficiency(proficiency_list):
    mapping = {
        "Fluent - Full Knowledge": "Fluent",
        "Fluent; speak, read, and write": "Fluent",
        "Fluent": "Fluent",
        "fluent": "Fluent",
        "Bi-lingual": "Bilingual",
        "Bilingual": "Bilingual",
        "mother tongue": "Native",
        "Native Speaker": "Native",
        "Native": "Native",
        "Professional working proficiency": "Professional",
        "Decreased proficiency due to moderate use": "Limited",
        "Knowledge, but taking classes to become more fluent": "Basic",
        "Fluent reading/writing": "Intermediate",
        "GCSE level": "Basic",
        "Good in both written and spoken": "Intermediate",
        "N/A": None,
        None: None
    }
    return [mapping.get(level, level) for level in proficiency_list if mapping.get(level, level) is not None]

resume_data["proficiency_levels"] = resume_data["proficiency_levels"].apply(clean_proficiency)
print(resume_data["proficiency_levels"].value_counts())
```

```python
# Merge languages with their proficiency levels
def merge_languages_proficiency(langs, profs):
    merged = []
    for i in range(len(langs)):
        level = profs[i] if i < len(profs) else "Unknown"  # Assign "Unknown" if missing
        merged.append(f"{langs[i]} ({level})")
    return ", ".join(merged)

resume_data["languages"] = resume_data.apply(lambda row: merge_languages_proficiency(row["languages"], row["proficiency_levels"]), axis=1)

# Drop old proficiency_levels column if no longer needed
resume_data.drop(columns=["proficiency_levels"], inplace=True)

# Show cleaned data
print(resume_data["languages"].head())
```

Extract City and State/Country:

If the address contains City, State (e.g., Chicago, IL) → Keep it as is.

If the address contains a full address (e.g., 2383 Suwanee Pointe Drive, Lawrenceville, GA 30043) → Extract City and State.

If the address contains a university department address (e.g., Department of Computer Science, University of British Columbia, Vancouver, BC, Canada) → Extract City and Country.

If the address contains a P.O. Box or generic placeholder (e.g., P.O BOX 2200 ELDORET) → Extract the city (Eldoret in this case).


If the address is ambiguous (e.g., Address, City, Province, Country) → Drop or infer from another column if possible.


Normalize formats:

Ensure consistency between state abbreviations and full names.

Convert country names to ISO country codes if needed.

Handle missing or generic values:

If an address is missing or too vague, try filling it from another column (e.g., education history or company name).

```python
import re
import pandas as pd

def extract_location(address):
    """Extracts city, state, or country from a given address."""
    if pd.isna(address) or 'Address' in address or 'City' in address:
        return None  # Skip generic placeholders

    # Regex for extracting City, State format (e.g., 'Chicago, IL' or 'Vancouver, BC')
    city_state_pattern = r'([A-Za-z\s]+),\s?([A-Z]{2})\b'
    match = re.search(city_state_pattern, address)
    if match:
        return f"{match.group(1).strip()}, {match.group(2).strip()}"

    # Extract full country name if present at the end
    country_pattern = r'([A-Za-z\s]+),\s?([A-Za-z\s]+),\s?([A-Za-z]+)$'
    country_match = re.search(country_pattern, address)
    if country_match:
        return f"{country_match.group(2).strip()}, {country_match.group(3).strip()}"

    # Extract just city if no state/country is found
    city_only_pattern = r'([A-Za-z\s]+)\s?\d*$'
    city_match = re.search(city_only_pattern, address)
    if city_match:
        return city_match.group(1).strip()

    return None  # Default case if no match

# Apply cleaning function to the dataset
resume_data['location'] = resume_data['address'].apply(extract_location)
resume_data.drop(columns=['address'])

# Display the first few cleaned locations
print(resume_data[['location']].value_counts())
```

```python
# تنظيف القيم في عمود degree_names
def clean_degree_names(value):
    if pd.isna(value):
        return "Unknown"  # استبدال القيم الفارغة
    
    if isinstance(value, str):
        try:
            # محاولة تحويل النص إلى قائمة إذا كان بتنسيق list
            value_list = ast.literal_eval(value)  
            if isinstance(value_list, list):
                # إزالة None وتحويل القائمة إلى نص مفصول بفواصل
                return ", ".join([item for item in value_list if item is not None])
        except (SyntaxError, ValueError):
            pass  # إذا فشل التحويل، نترك القيمة كما هي

    return value  # إعادة القيمة الأصلية إذا لم تكن بحاجة إلى تعديل

# تطبيق الدالة على العمود
resume_data["degree_names"] = resume_data["degree_names"].apply(clean_degree_names)

# التحقق من النتيجة بعد التنظيف
print(resume_data["degree_names"].value_counts())
```

Convert Strings to Lists Properly:

Some values are single dates ('Nov 2019'), while others are lists ("['03/2014', '06/2003', '01/2002']").
We need to extract dates correctly from string representations of lists.
Standardize Date Formats:

Some dates are in "MMM YYYY" format (e.g., 'Nov 2019'), while others are in "MM/YYYY" format (e.g., '03/2014').
We'll parse both formats correctly.
Extract Minimum Date for Each Row:

If a row contains multiple dates, we'll find the earliest (oldest) date.
Handle Missing or Invalid Data:

Convert any NaN or empty lists to NaT (Not a Time).
Handle errors gracefully during conversion.

```python
import pandas as pd
import ast  # To safely convert string representations of lists

def extract_min_date(date_list):
    if pd.isna(date_list) or date_list == "[]":
        return pd.NaT  # Return NaT if empty or NaN
    
    if isinstance(date_list, str):  
        try:
            date_list = ast.literal_eval(date_list)  # Convert string to list
        except (ValueError, SyntaxError):
            date_list = [date_list]  # If it's a single string, treat it as a list
    
    if not isinstance(date_list, list):
        date_list = [date_list]  # Ensure it's a list
    
    parsed_dates = []
    for date in date_list:
        date = str(date).strip()
        
        # Try parsing different date formats
        try:
            if "/" in date:  # Format: MM/YYYY
                parsed_dates.append(pd.to_datetime(date, format="%m/%Y", errors="coerce"))
            else:  # Format: MMM YYYY (e.g., 'Nov 2019')
                parsed_dates.append(pd.to_datetime(date, format="%b %Y", errors="coerce"))
        except Exception:
            continue  # Skip invalid dates
    
    # Return the minimum valid date
    return min(parsed_dates) if parsed_dates else pd.NaT

# Apply the function to the column
resume_data["start_dates"] = resume_data["start_dates"].apply(extract_min_date)

# Display cleaned data
resume_data.start_dates.value_counts()
```

Steps for Data Cleaning
Convert Strings to Lists Properly:

Some values are single dates ('Oct 2020'), while others are lists ("['10/2019', '11/2013', '01/2003']").
We need to extract dates correctly from string representations of lists.
Standardize Date Formats:

Some dates are in "MMM YYYY" format (e.g., 'Oct 2020'), while others are in "MM/YYYY" format (e.g., '10/2019').
We'll parse both formats correctly.
Handle "Current" Values:

Values like 'Till Date', 'Ongoing', 'Current', and variations like 'till date' should be converted to NaT (Not a Time) or today's date.
Extract Maximum Date for Each Row:

If a row contains multiple dates, we'll find the latest (most recent) date.
Handle Missing or Invalid Data:

Convert any NaN or empty lists to NaT.
Handle errors gracefully during conversion.

```python
import pandas as pd
import ast  # To safely convert string representations of lists
from datetime import datetime

def extract_max_date(date_list):
    if pd.isna(date_list) or date_list == "[]":
        return pd.NaT  # Return NaT if empty or NaN
    
    if isinstance(date_list, str):  
        try:
            date_list = ast.literal_eval(date_list)  # Convert string to list
        except (ValueError, SyntaxError):
            date_list = [date_list]  # If it's a single string, treat it as a list
    
    if not isinstance(date_list, list):
        date_list = [date_list]  # Ensure it's a list
    
    parsed_dates = []
    for date in date_list:
        date = str(date).strip()

        # Handle 'Current' values
        if date.lower() in ['till date', 'ongoing', 'current']:
            parsed_dates.append(pd.Timestamp.today())  # Use today's date

        else:
            # Try parsing different date formats
            try:
                if "/" in date:  # Format: MM/YYYY
                    parsed_dates.append(pd.to_datetime(date, format="%m/%Y", errors="coerce"))
                else:  # Format: MMM YYYY (e.g., 'Oct 2020')
                    parsed_dates.append(pd.to_datetime(date, format="%b %Y", errors="coerce"))
            except Exception:
                continue  # Skip invalid dates
    
    # Return the maximum valid date
    return max(parsed_dates) if parsed_dates else pd.NaT

# Apply the function to the column
resume_data["end_dates"] = resume_data["end_dates"].apply(extract_max_date)

# Display cleaned data
resume_data.end_dates.value_counts()
```

```python
import pandas as pd

def clean_responsibilities(text):
    if pd.isna(text):  # Handle missing values
        return "Unknown"
    text = text.strip()  # Remove leading/trailing whitespace
    text = text.replace("\n", ", ")  # Convert newlines to commas
    text = ', '.join(set(text.split(", ")))  # Remove duplicate words
    text = text.title()  # Standardize capitalization
    return text

# Apply cleaning function
resume_data["responsibilities"] = resume_data["responsibilities"].apply(clean_responsibilities)

# Display cleaned data
print(resume_data["responsibilities"].head(10))
```

```python
# Function to safely convert string lists to actual lists
def safe_eval(val):
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)  # Convert string to list
        except (ValueError, SyntaxError):
            return []
    return val if isinstance(val, list) else []

# Apply safe conversion
resume_data['role_positions'] = resume_data['role_positions'].apply(safe_eval)
resume_data['positions'] = resume_data['positions'].apply(safe_eval)

# Function to clean and merge work history
def merge_work_history(row):
    roles = [r for r in row['role_positions'] if r and r != 'N/A' and r != 'None']
    positions = [p for p in row['positions'] if p and p != 'N/A' and p != 'None']
    responsibilities = row['responsibilities'] if isinstance(row['responsibilities'], str) else ""

    work_history = []
    if roles:
        work_history.append(f"Roles: {', '.join(roles)}")
    if positions:
        work_history.append(f"Positions: {', '.join(positions)}")
    if responsibilities:
        work_history.append(f"Responsibilities: {responsibilities}")

    return ' | '.join(work_history) if work_history else 'No work history available'

# Apply cleaning and merging
resume_data['work_history'] = resume_data.apply(merge_work_history, axis=1)

resume_data.work_history.value_counts()
```
